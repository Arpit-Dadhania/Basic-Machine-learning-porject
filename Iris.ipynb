```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.datasets import load_iris
```

```python
iris=load_iris(as_frame=True)
df=iris.frame
df.head()
```

```python
df.shape
```

```python
sns.scatterplot(df,x="sepal length (cm)",y="sepal width (cm)",hue="target")
```

```python
sns.scatterplot(df,x="petal length (cm)",y="petal width (cm)",hue="target")
```

```python
df_1=df.drop("target",axis=1)
df_1.head()
```

```python
df_2=df["target"]
df_2.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_1,df_2,test_size=0.2,random_state=42)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test.tolist()
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

```python
cf=confusion_matrix(y_test,y_pred)
cf
```

```python
cf=confusion_matrix(y_test,y_pred)
cf
```

```python
label=["Iris setosa","Iris versicolor" ,"Iris virginica" ]
sns.heatmap(cf,annot=True,xticklabels=label,yticklabels=label)
plt.show()
```
